```python
import hashlib
from ecdsa import SigningKey, SECP256k1

#Transaction message
message = "Ahmed pays 100 coins to Nedaa"
print("Transaction Message:", message)

#Hash the message using SHA-256

message_hash = hashlib.sha256(message.encode()).hexdigest()
print("SHA-256 Hash:", message_hash)

#Generate ECDSA privateâ€“public key pair
private_key = SigningKey.generate(curve=SECP256k1)
public_key = private_key.get_verifying_key()

#Sign the hash using the private key
signature = private_key.sign(message_hash.encode()).hex()
print("Signature:", signature)

#Verify the signature using the public key
try:
    verification = public_key.verify(bytes.fromhex(signature), message_hash.encode())
    print("Signature Verified:", True)
except:
    print("Signature Verified:", False)
```
