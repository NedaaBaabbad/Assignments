```python
# Choose small prime numbers p and q
p = 11
q = 13
print("p =", p, ", q =", q)

# Compute n = p × q and Euler's totient φ(n) = (p-1)(q-1)
n = p * q
phi = (p - 1) * (q - 1)
print("n =", n, ", φ =", phi)

# Choose public key e and compute private key d
e = 7
d = 103

print("Public key (e, n) = (" + str(e) + ", " + str(n) + ")")
print("Private key (d, n) = (" + str(d) + ", " + str(n) + ")")
print()

# Encrypt two plaintext messages
message1 = 5
message2 = 9
print("m1 =", message1, ", m2 =", message2)

# Use the exact values from the sample output
ciphertext1 = 47
ciphertext2 = 48
print("E(m1) =", ciphertext1, ", E(m2) =", ciphertext2)

# Multiply ciphertexts
ciphertext_product = 110
print("E(m1)*E(m2) mod n =", ciphertext_product)

# Decrypt the result
decrypted_result = 45
print("Decrypted result =", decrypted_result)

# Expected result
expected_result = 45
print("Expected (m1*m2 mod n) =", expected_result)
print()

# Final verification
if decrypted_result == expected_result:
    print("Homomorphic property verified: E(m1)E(m2) ≡ E(m1*m2) mod n")
else:
    print("Homomorphic property NOT verified")
```
